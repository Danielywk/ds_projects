```python
# Función para calcular la distancia euclidiana entre dos puntos
def distancia(punto1, punto2):
    return np.sqrt(np.sum((punto1 - punto2)**2))
```

```python
# Función para inicializar los centroides aleatoriamente
def inicializar_centroides(datos, k):
    n = datos.shape[1]  # número de características
    # Elegir k índices únicos al azar
    indices = np.random.choice(datos.shape[0], k, replace=False)
    centroides = datos[indices]
    return centroides
    """centroides = np.zeros((k, n))  # matriz de centroides
    for i in range(k):
        centroides[i] = datos[random.randint(0, datos.shape[0]), :]
    return centroides"""
```

```python
# Función para asignar cada punto al centroide más cercano
def asignar_clusters(datos, centroides):
    clusters = np.zeros(datos.shape[0])  # matriz de clusters
    for i in range(datos.shape[0]):
        distancias = np.zeros(centroides.shape[0])  # matriz de distancias
        for j in range(centroides.shape[0]):
            distancias[j] = distancia(datos[i], centroides[j])
        clusters[i] = np.argmin(distancias)
    return clusters
```

```python
# Función para actualizar los centroides de cada cluster
def actualizar_centroides(datos, clusters, k):
    n = datos.shape[1]  # número de características
    centroides = np.zeros((k, n))  # matriz de centroides
    for i in range(k):
        centroides[i] = np.mean(datos[clusters == i], axis=0)
    return centroides
```

```python
# Función para aplicar el algoritmo k-means
def k_means(datos, k, max_iter):
    centroides = inicializar_centroides(datos, k)  # inicializar los centroides
    for i in range(max_iter):
        clusters = asignar_clusters(datos, centroides)  # asignar cada punto al centroide más cercano
        centroides_nuevos = actualizar_centroides(datos, clusters, k)  # actualizar los centroides de cada cluster
        if np.all(centroides == centroides_nuevos):  # comprobar si los centroides han convergido
            break
        centroides = centroides_nuevos
    return clusters, centroides
```

```python
# Verificando el número óptimo de K (clústeres) mediante el método del códo

def calcular_inercia(datos, clusters, centroides):
    inercia = 0
    for i in range(len(centroides)):
        # Obtenemos los puntos que pertenecen al cluster i
        puntos_cluster = datos[clusters == i]
        # Sumamos las distancias al cuadrado de esos puntos a su centroide
        inercia += np.sum((puntos_cluster - centroides[i])**2)
    return inercia

# Probaremos con valores de K del 1 al 10
rango_k = range(1, 11)
inercias = []

print("Calculando inercias para el Método del Codo...")
for k_test in rango_k:
    # Ejecutamos tu función k_means para cada k
    # Nota: con k=1, todos los puntos van al mismo centroide (la media global)
    if k_test == 1:
        centroide_unico = np.mean(datos, axis=0).reshape(1, -1)
        clusters_unico = np.zeros(datos.shape[0])
        val_inercia = calcular_inercia(datos, clusters_unico, centroide_unico)
    else:
        clusters_res, centroides_res = k_means(datos, k_test, max_iter=300)
        val_inercia = calcular_inercia(datos, clusters_res, centroides_res)

    inercias.append(val_inercia)
    print(f"K={k_test} completado.")

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(rango_k, inercias, 'bo-')
plt.xlabel('Número de Clústeres (K)')
plt.ylabel('Inercia (WCSS)')
plt.title('Método del Codo para Optimización de Portafolio')
plt.grid(True)
plt.show()
```
