```python
# IMPORTAR LIBRERÍAS
import pandas as pd
import numpy as np
from statistics import mean
import random
import matplotlib.pyplot as plt
import io
```

```python
# CARGANDO LA DATA
datos = pd.read_csv("datasources/s22_sit.csv")
print(datos)
```

```python
# SELECCIÓN DE LOS 2000 DATOS
datos1 = datos.loc[2001:4000]
datos1 = datos1.iloc[:,1]
print   (datos1)
datos11 = datos1.loc[2001:4000]
```

```python
# GRAFICACIÓN DEL ELECTROCARD
#Electrocardiograma inicial
datos1.plot(y = 'ecg', figsize=(28,3), xlabel='Tiempo (seg)', ylabel='ECG', title='Electrocardiograma')
```

```python
# SUTITUCIÓN NA
sust=datos11.sample(20)
new = list(sust)
print (new)

# GENERANDO NUEVA VAR
Nvar=datos11
for x in new:
  Nvar.loc[Nvar== x ] = None
  print (Nvar)
```

```python
# GRAFICANDO DATOS DE LA NUEVA VARABLE
Nvar.plot( y='ecg', figsize=(28,3), xlabel='Tiempo', ylabel='ECG',title='Graficación con NA')
plt.show()
```

```python
# COPIA Y SUSTITUCIÓN DE LAS IMPUTACIONES
datosNA = datos11.copy()
for x in new:
  datosNA.loc[datosNA== x ] = None
```

```python
# IMPUTACIONES
datosMean = datos11.copy()
datosMediana = datos11.copy()
moda = datosNA.copy()
datosInterO1 = datos11.copy()
datosInterO2= datos11.copy()
datos0=datosNA.copy()
datosMedMov=datos11.copy()
print (datosNA)
```

```python
# GRÁFICA DE NÚMEROS RAND EN LA LISTA DE REEMPLAZO
mayor=datosNA.max().max()
menor=datosNA.min().min()
print(mayor , menor)
remp= datosNA.isna().sum()
listaRemp = np.random.randint(low=menor, high=mayor, size=(remp))
listaRemp
```

```python
for x in listaRemp:
  datosNA.fillna(x,inplace=True)
  datosNA
  print(datosNA)
```

```python
# GRAFICANDO ELECTRO CON LOS NÚMEROS ALEA
datosNA.plot( y='ecg', figsize=(28,3), xlabel='Tiempo', ylabel='ECG',title='Gráfica de sustitución NA')
plt.show()

# ELECTRO ORIGINAL
Nvar.plot( y='ecg', figsize=(28,3), xlabel='Tiempo', ylabel='ECG',title='ELECTROCARDIOGRAMA INICIAL')
plt.show()
```

```python
# OBTENIENDO LA MEDIANA
Median = datosMediana.median()
datosMediana.fillna(Median)
# REEMPLAZO
datosMediana.plot( y='ecg', figsize=(20,6), xlabel='Tiempo', ylabel='ECG',title='Sustitución NA con Median')
plt.show()
```

```python
# OBTENIENDO EL PROMEDIO
Mean = datosMean.mean()
datosMean.fillna(Mean)
# REEMPLAZO
datosMean.plot( y='ecg', figsize=(28,3), xlabel='Tiempo', ylabel='ECG',title='Sustitución NA con Mean')
plt.show()
```

```python
# INTERPOLACION
INTERO1 = datosInterO1.interpolate()
datosInterO1.fillna(INTERO1)
#REMPLAZO
datosInterO1.plot( y='ecg', figsize=(28,3), xlabel='Tiempo', ylabel='ECG',title= "INTERPOLACIÓN ORDÉN 1")
```
